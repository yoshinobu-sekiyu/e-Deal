# 4. Server-side scripting

```
serverpy = """
import tornado.web
import tornado.httpserver
import tornado.ioloop

with open("./template/motorhome.html", "r") as f:

  motorhome_html = f.read()

motorhome_html.replace("#woodpanel", "#woodboard")


class MotorApp(tornado.web.Application):
    
    def __init__(self):
      
        handlers:List = [

            (r"/", MainHandler),
                       
        ]
        
        tornado.web.Application.__init__(self,handlers,xsrf_cookies= True)
        
class MainHandler(tornado.web.RequestHandler):
    

    def get(self) -> None:
        
        self.write(motorhome_html)


def main():
    
    app = MotorApp()
    http_server = tornado.httpserver.HTTPServer(app)
    http_server.listen(8890)
    tornado.ioloop.IOLoop.current().start()

if __name__ == "__main__":
  main()
"""


with open("edeal/server.py", "w") as f:

  f.write(serverpy)
```
