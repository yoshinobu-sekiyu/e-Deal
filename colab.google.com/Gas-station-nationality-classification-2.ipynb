```python
from pathlib import Path
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = Path("../datasets/raw")

paths = list(p.glob("**/*.png"))
paths
```

```python
# Vizualizing multiple images at once.

fig=plt.figure(figsize=(8, 8))
columns = 3
rows = 5
for i in range(1, columns*rows +1):
    img = imread(paths[i])
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```

```python
# Resizing images.

import matplotlib.pyplot as plt
from skimage import color
from skimage.transform import rescale, resize, downscale_local_mean
import numpy as np
from sympy import *

img0 = imread(paths[0])
image = color.rgb2gray(img0)

ratio = image.shape[0]/ image.shape[1]

image_resized = resize(image, (image.shape[0] , image.shape[1] * ratio),
                       anti_aliasing=True)

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,6))

ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original image")

ax[1].imshow(image_resized, cmap='gray')
ax[1].set_title("Rescaled image (aliasing)")

ax[0].set_xlim(0, 1024)
ax[0].set_ylim(1024, 0)
plt.tight_layout()

plt.show()
```

```python
# Sympy can also be used but algebraic-ratio seems to be too strict when computation.

x = Symbol("x")
y = Symbol("y")

x_im = image.shape[0]/x
y_im = image.shape[1]/y

sln = solve(x_im - y_im)
sln
```

```python
fig=plt.figure(figsize=(8, 8))
columns = 3
rows = 5
for i in range(1, columns*rows +1):
    img = imread(paths[i])
    ratio = img.shape[0]/ img.shape[1]
    img_resized = resize(img, (img.shape[0] , img.shape[1] * ratio),
                       anti_aliasing=True)

    fig.add_subplot(rows, columns, i)
    plt.imshow(img_resized)
plt.show()
```
